```python
import pandas as pd
from asag2024_dataset.datasets.digiklausur_loader import load_digiklausur, rename_digiklausur
import nltk
```

```python
df = rename_digiklausur(load_digiklausur())
df.head()
```

```python
df.info()
```

```python
calculate_word_count = lambda answer: len(nltk.word_tokenize(str(answer)))
calculate_sentence_count = lambda answer: len(nltk.sent_tokenize(str(answer)))

df['sentence_count_provided_answer'] = df['provided_answer'].apply(calculate_sentence_count)
df['word_count_provided_answer'] = df['provided_answer'].apply(calculate_word_count)

df['sentence_count_reference_answer'] = df['reference_answer'].apply(calculate_sentence_count)
df['word_count_reference_answer'] = df['reference_answer'].apply(calculate_word_count)
```

```python
number_of_na_answers = len(df[pd.isna(df["provided_answer"])])
number_of_na_answers
```
